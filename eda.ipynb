```python
import pandas as pd
import datatable as dt
from utility.plot_utility import KernelPlot
from datatable import f
```

```python
data_docs = dt.fread('data/unctad_document_level_data.csv')
data_arbs = dt.fread('data/arbitration_level_document_metadata.csv')

data_docs = dt.Frame(data_docs)
data_arbs = dt.Frame(data_arbs)
```

```python

data_docs = data_docs[f.page_count != None, :]
data_docs = data_docs.to_pandas()

data_arbs = data_arbs[f.page_sum != None, :]
data_arbs = data_arbs.to_pandas()
```

```python
kde_docs = KernelPlot(data_docs, 'page_count', 'Page Count')
kde_docs.plot_density()
```

```python
kde_arbs = KernelPlot(data_arbs, 'page_sum', 'Page Sum')
kde_arbs.plot_density()
```
